```python
import os

import numpy as np
from matplotlib import pyplot as plt

from main import parse_snippet, get_snippet_event
```

```python
def plot_snippet(time_slice: np.ndarray, signal_slice: np.ndarray, title: str = ""):
    plt.plot(time_slice, signal_slice)
    plt.xlabel("Time (s)")
    plt.ylabel("Normalised amplitude")
    plt.ylim([-1.2, 1.2])
    plt.title(title)
```

```python
snippet_folder = "Snippets"
snippet_file = "Elsa_5moderateblinks_5moderateblinks_1.npy"
snippet = np.load(f"{snippet_folder}/{snippet_file}")
signal_slice, time_slice = parse_snippet(snippet)
```

```python
plot_snippet(time_slice, signal_slice)
```
