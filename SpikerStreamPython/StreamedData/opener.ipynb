```python
import numpy as np
```

```python
arr = np.load("streamed_noise_custom.npy")
```

```python
from matplotlib import pyplot as plt
plt.plot(arr);
```

```python
from astropy import units as u
from astropy.convolution import convolve, Box1DKernel
smooth_width = len(arr) // 50
smooth_amplitude = convolve(arr, Box1DKernel(smooth_width))
plt.plot(smooth_amplitude);
```
